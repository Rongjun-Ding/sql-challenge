```python
salary = pd.read_csv('Resources/data1.csv')
salary
```

```python
# Figure out the minimum and maximum views for salary
print(salary["salary"].max())
print(salary["salary"].min())
```

```python
# Create bins in which to place values based upon salary
bins = [0, 50000, 100000, 150000,]

# Create labels for these bins
group_labels = ["0 to 50k", "50k to 100k", "100k to 150k"]
```

```python
# Place the data series into a new column inside of the DataFrame
salary["common salary ranges"] = pd.cut(salary["salary"], bins, labels=group_labels)
salary.head()
```

```python
# Create a GroupBy object based upon "common salary ranges"
salary = salary.groupby("common salary ranges")

# Find how many rows fall into each bin
print(salary["common salary ranges"].count())
```

```python
# Get the average of each column within the GroupBy object
A= salary["salary"].mean()
A
```
